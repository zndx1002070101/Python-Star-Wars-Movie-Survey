## Read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

## Cleaning and Mapping Yes/No Columns

```python
map_1 = {'Yes': True, 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(map_1)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(map_1)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

```python
# using mapping

for col in star_wars.columns[3:9]:
    map_col = {star_wars[col][0]: True, np.NaN: False}
    star_wars[col] = star_wars[col].map(map_col)
```

```python
star_wars.head()
```

```python
columns_rename = {star_wars.columns[3]:'seen_1', star_wars.columns[4]:'seen_2', star_wars.columns[5]:'seen_3',
                 star_wars.columns[6]:'seen_4', star_wars.columns[7]:'seen_5', star_wars.columns[8]:'seen_6'}
star_wars = star_wars.rename(columns=columns_rename)
star_wars.head()
```

## Cleaning the Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rank_columns_rename = {star_wars.columns[9]:'rank_1', star_wars.columns[10]:'rank_2', star_wars.columns[11]:'rank_3',
                 star_wars.columns[12]:'rank_4', star_wars.columns[13]:'rank_5', star_wars.columns[14]:'rank_6'}
```

```python
star_wars = star_wars.rename(columns=rank_columns_rename)
star_wars.head()
```

## Finding the Highest-Ranked Movie

```python
# mean rank of each movie
star_wars.iloc[:, 9:15].mean().plot.bar()
plt.ylabel('Mean rank')
plt.show()
```

Star Wars: Episode V The Empire Strikes Back is most favored by the respondents.

## Finding the Most Viewed M2ovie

```python
# Most viewed movie
%matplotlib inline
star_wars.iloc[:, 3:9].sum().plot.bar()
plt.ylabel('Total views')
plt.show()
```

Star Wars: Episode V The Empire Strikes Back is most viewed movie by the respondents.It suggests that people like Episode V.

## Exploring the Data by Binary Segments

```python
# View by sex
star_wars.groupby('Gender')[['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']].sum().plot.bar()
plt.ylabel('Total views')
plt.show()
```

Both female and male likes to watch Episode V.

```python

# Rank by sex, the lower the rank, the better the movie
star_wars.groupby('Gender')[['rank_1', 'rank_2', 'rank_3', 'rank_4', 'rank_5', 'rank_6']].mean().plot.bar()
plt.ylabel('Mean rank')
plt.show()
```

Both female and male give a high rating of Episode V.

```python
# Rank, by fans
star_wars.groupby('Do you consider yourself to be a fan of the Star Wars film franchise?')[['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']].sum().plot.bar()
plt.ylabel('Number of people')
plt.show()
```

Among the fans or non-fans, most of the people are fans of Episode V.

```python
star_wars.groupby('Do you consider yourself to be a fan of the Star Wars film franchise?')[['rank_1', 'rank_2', 'rank_3', 'rank_4', 'rank_5', 'rank_6']].mean().plot.bar()
plt.ylabel('Mean rank')

plt.show()
```

Among the fans or non-fans, the rating of Episode V is the highest.
